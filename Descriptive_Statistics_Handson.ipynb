Here, in pandas we don't have any methods to calculate percentile. But in Numpy we do have the above method for percentile. Here, 25 means 25th percentile of data, Simply N = Nth percentile of data.

```python
# Quartile
Q1 = df_country['Population in thousands (2017)'].quantile(0.25)
Q2 = df_country['Population in thousands (2017)'].quantile(0.50)
Q3 = df_country['Population in thousands (2017)'].quantile(0.75)

print(Q1, Q2, Q3)
```

Here, the method is 'quantile' for calculating percentile. We given 0.25 which is taken as percentage (25% quantile) to compute and value always should be " 0 <= value => 1 ". It means that 25% of data found under this value while arranged in ascending order.

```python
# Inter-Quartile Range
IQR = Q3 - Q1
IQR
```

```python
# Lower limit
LL = Q1 - 1.5 * IQR
LL
```

```python
# Upper limit
UL = Q3 + 1.5 * IQR
UL
```

### Sampling

```python
# Random Sampling
sample1 = df_country['Population in thousands (2017)'].sample(100)
sample1
```

Here as we see that this sampling takes random data every time while execution, thus it is random sampling.

Now, we take 'Region' column for stratified sampling, and we'll take 70% percent of data from each 'region'.

```python
df_country['Region'].value_counts()
```

```python
# Display all the rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

Note :- 
- The above method is used to display every rows & columns no matter how long the data is, it will display all of them.

```python
# Stratified sampling (with group_keys = False)
sample_region = df_country.groupby('Region', group_keys=False).apply(lambda x : x.sample(frac=0.7)) # frac means fraction
sample_region                                                                                       # 0.7 means 70 %
```

Note :- 
- We cannot divide the number of records equally for all the categories, cause each categories have different number of records.<br/>
  But we can give percentage. Although the number may/may not be same but the contribution from each category will be same.<br/> 
  So we are not having any biasness, and we are able to equally extract the records.

Simply above statement means "Whenever we are using stratified sampling, we cannot give sample size directly in number instead we give it in fraction (0.0) which will taken as percentage (%) by kernel".<br/>
i.e., we passed '0.7' which taken as 70%.

```python
# Stratified sampling (with group_keys = True)
sample_region = df_country.groupby('Region', group_keys=True).apply(lambda x : x.sample(frac=0.7)) # frac means fraction
sample_region                                                                                      # 0.7 mean 70% 
```

Note :- 
- Now, for every 'Region'(category used for grouping) the function "group_keys" will create an additional key which shows region too but in different manner.<br/> <br/>
- like if the additional key REGION(created by group_keys) is showing 'carribean' and beside that there will be all the records comes in region of carribean.

**Group_keys** :- <br/>
This function is used to create an additional keys for the respective category we used for grouping the data, <br>
works just like index such that it show the CATEGORY(additional key) and besides that <br/>
all records will be shown which falls under that category.
